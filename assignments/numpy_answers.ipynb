```python
ar[2:5,-2:]
```

```python
ar[::2,-1:] # This will produce a Matrix (3x1)
```

```python
ar[::2,-1]  # This will give a Vector
```

#### Display highest value in each row in 5,5 array 

```python
a = np.random.randint(1,50,(5,5))
print(a)
max_avg = a.max(axis=1)
print("Highest value in each row = ", max_avg)
```

#### Display average of each row 

```python
avg = a.mean(axis=1)
print("Average of each row = ", avg)
```

<h2 style="color:red">Reshaping, stacking and splitting an Array</h2>

#### Reshape an array of 2,6 to 3,4 

```python
a = np.random.randint(1,10,(2,6))
print(a)
a.reshape((3,4))
```

#### Convert a 3,3 array to a single dimension array and create a copy

```python
a = np.random.randint(1,10,(3,3))
print(a)
b = a.flatten()
b
```

#### Reshape an array of 2,6 to 3,4 inplace

```python
a = np.random.randint(1,10,(2,6))
print(a)
a.resize((3,4))
a
```

#### Create an array of 5,5 and transpose it 


```python
a = np.random.randint(1,25,(5,5))
print(a)
a.T
```

#### Split an array of 5,5 into three parts at 2nd and 4th rows 

```python
np.vsplit(a,(2,4))
```

#### Concatenate two arrays of 3,3 and 4,3 into one array

```python
a = np.random.randint(1,20,(3,3))
print(a)
b = np.random.randint(1,20,(4,3))
print(b)
np.vstack((a,b))
```

#### Split an array of 6,6 into two parts horizontally

```python
a = np.random.randint(10,50,(6,6))
print(a)
np.hsplit(a,2)
```
