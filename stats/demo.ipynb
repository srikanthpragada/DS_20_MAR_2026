#### Skewness 

```python
#Normal distribution  - skewness is 0
marks = [10,20,30,40,50,60,70,80]
ss.skew(marks)
```

```python
#Skewed to right - positive skewness 
marks = [10,20,30,40,50,60,70,120,200]
print(ss.skew(marks), s.mean(marks), s.median(marks))
```

```python
#Skewed to left - - Negative skewness 
marks = [10,100,110,120,150]
ss.skew(marks)
```

#### kurtosis

```python
# Negative kurtosis - Less data in tails 
marks = [60,70,80,90,100]
ss.kurtosis(marks)
```

```python
# Negative kurtosis - Less data in tails 
marks = [10,20,30,40,50,60,70,80]
ss.kurtosis(marks)
```

```python
# Positive krutosis - More data in tails 
marks = [10,15,20,90,90,90,100,150,200,300,600]
ss.kurtosis(marks)
```

### Correlation

```python
gre = [320,311,290,299,330,280,260,298,310,315]
cgpa = [8,9,7,7.5,8,8.5,7.2,8.3,9.1,9.2]
ss.pearsonr(cgpa,gre)
```

### Linear Regression

```python
slope, intercept, r_value, p_value, std_err = ss.linregress(cgpa,gre)
print(f"Intercept = {intercept}, Slope = {slope}")
x = 8.7 # CGPA (Independent variable)
y = intercept + slope * x
print(f"We predict GRE score of {y} for CGPA of {x}")
```
