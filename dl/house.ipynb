```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
```

```python
# Input data - SQFT, BED Rooms, Age in Years
X = np.array([
    [1000, 2, 10],
    [1500, 3, 5],
    [800, 2, 20],
    [2000, 4, 2]
], dtype=float)

y = np.array([2000000, 3000000, 1500000, 4000000], dtype=float)
```

```python
X.max(axis=0)
```

```python
# Normalize data (important for neural networks)
X_max = X.max(axis=0)
y_max = y.max()

X = X / X_max
y = y / y_max
```

```python
X
```

```python
# Build model
model = keras.Sequential([
    keras.Input(shape=(3,)),
    layers.Dense(2, activation='relu', ),  
    layers.Dense(1, activation='linear')    # Output layer
])
```

```python
model.summary()
```

```python
# Compile
model.compile(optimizer='adam', loss='mse')
```

```python
# Train
model.fit(X, y, epochs=50)
```

```python
y_p = model.predict(X) * y_max
```

```python
y_p
```

```python
model.get_weights()
```

```python
y_p[:,0] - (y * y_max)
```

```python
# Predict
test = np.array([[1200, 3, 8]])
test = test / X_max
```

```python
prediction = model.predict(test)
print(prediction * y_max)
```
